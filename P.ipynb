```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
try:
    data = pd.read_csv(r'/content/car_price_dataset.csv', encoding='ISO-8859-1')
    print('File load: Success')
except:
    print('File load: Failed')
```

```python

print(data)
```

```python
data.shape
```

```python
no_of_cols = data.columns.nunique()
print(no_of_cols)
```

```python
data = data.rename(columns={'Unnamed: 0': 'SR.NO'})
```

```python
data
```

```python
data = data[['Year','Engine_Size','Fuel_Type','Transmission','Mileage','Owner_Count','Price']]
```

```python
data
```

```python
print(data.isnull().sum())

```

```python
plt.figure(figsize=(20,5));
sns.countplot(x=data.Engine_Size);
```

```python
plt.figure(figsize=(20, 6))
sns.countplot(x='Year', data=data)
plt.xticks(rotation=45)
plt.title('Sales by Year')
plt.show()
```

```python
grouped_data = data.groupby('Year').sum()
plt.pie(grouped_data['Mileage'], labels=grouped_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart from Grouped Data')
plt.show()
```

```python
grouped_data = data.groupby('Year').sum()
plt.pie(grouped_data['Price'], labels=grouped_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart from Grouped Data')
plt.show()
```
